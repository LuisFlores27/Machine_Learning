# 1. Fundamentos de la técnica SLR

La **Regresión Lineal Simple (Simple Linear Regression, SLR)** es un algoritmo de aprendizaje supervisado utilizado para modelar la relación entre dos variables:

- **Variable independiente (X):** variable que se usa para hacer la predicción.
- **Variable dependiente (Y):** variable que se desea predecir.

El objetivo del algoritmo es encontrar una **línea recta** que represente la relación entre ambas variables.

Esta línea se utiliza para:
- analizar la relación entre variables
- hacer predicciones de valores futuros
- entender tendencias en los datos

La regresión lineal es uno de los modelos más básicos y utilizados en **Machine Learning y análisis de datos**.

# 2. Modelo Matemático

La ecuación de la regresión lineal simple es:

y = mx + b

donde:

- **y** → variable dependiente (valor que queremos predecir)
- **x** → variable independiente
- **m** → pendiente de la recta
- **b** → intercepto (valor donde la recta cruza el eje Y)

El algoritmo busca encontrar los valores óptimos de **m** y **b** que minimicen el error entre los valores reales y los valores predichos.

Este proceso se realiza mediante el método de **mínimos cuadrados**.

# 3. Implementación paso a paso

## Paso 1: Importar librerías

Antes de implementar el modelo es necesario importar las librerías necesarias para trabajar con datos, visualización y machine learning.

Las librerías utilizadas son:

- **numpy** → operaciones matemáticas y manejo de arreglos
- **pandas** → manipulación de datos
- **matplotlib** → visualización de datos
- **scikit-learn** → implementación de algoritmos de Machine Learning

```python
# numpy se utiliza para trabajar con arreglos y operaciones matemáticas
import numpy as np

# pandas permite cargar y manipular datasets
import pandas as pd

# matplotlib permite crear gráficos
import matplotlib.pyplot as plt

# train_test_split permite dividir los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split

# LinearRegression es la clase que implementa el modelo de regresión lineal
from sklearn.linear_model import LinearRegression
```

## Paso 2: Crear el conjunto de datos

Para este ejemplo se utilizarán datos simples que representan la relación entre:

- inversión en publicidad
- ventas obtenidas

Se almacenan en dos listas y posteriormente se convierten en arreglos de numpy.

```python
# Variable independiente (por ejemplo inversión en publicidad)
X = np.array([23, 26, 30, 34, 43, 48, 52, 57, 58]).reshape(-1,1)

# Variable dependiente (por ejemplo ventas)
y = np.array([651, 762, 856, 1063, 1190, 1298, 1421, 1440, 1518])
```

## Paso 3: Dividir los datos en entrenamiento y prueba

Para evaluar el modelo correctamente, los datos se dividen en:

- **datos de entrenamiento** → utilizados para entrenar el modelo
- **datos de prueba** → utilizados para evaluar el modelo

La función `train_test_split()` de scikit-learn permite hacer esta división automáticamente.

```python
# División de los datos:
# test_size=0.2 indica que el 20% de los datos serán para prueba
# random_state asegura que la división sea reproducible

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

## Paso 4: Crear el modelo de regresión lineal

La librería scikit-learn proporciona la clase **LinearRegression** para crear modelos de regresión lineal.

En este paso se crea un objeto que representará el modelo.

```python
# Se crea una instancia del modelo de regresión lineal
regressor = LinearRegression()
```

## Paso 5: Entrenar el modelo

El método `fit()` se utiliza para entrenar el modelo.

Durante el entrenamiento el algoritmo calcula:
- la pendiente (m)
- el intercepto (b)

Estos valores permiten construir la línea de regresión.

```python
# El modelo se entrena utilizando los datos de entrenamiento
regressor.fit(X_train, y_train)
```

## Paso 6: Realizar predicciones

Una vez entrenado el modelo, se puede utilizar el método `predict()` para estimar valores de salida.

En este caso se utilizarán los datos de prueba para evaluar el modelo.

```python
# Predicción de los valores usando los datos de prueba
y_pred = regressor.predict(X_test)
```

## Paso 7: Visualización de resultados

Para entender mejor el comportamiento del modelo se grafican:

- los datos originales
- la línea de regresión generada por el modelo

Esto permite observar qué tan bien se ajusta la recta a los datos.

```python
# Gráfico de los datos de entrenamiento
plt.scatter(X_train, y_train, color='red')

# Línea de regresión generada por el modelo
plt.plot(X_train, regressor.predict(X_train), color='blue')

# Título del gráfico
plt.title("Regresión Lineal Simple")

# Etiquetas de los ejes
plt.xlabel("Publicidad")
plt.ylabel("Ventas")

# Mostrar el gráfico
plt.show()
```
